```
import numpy as np
import matplotlib.pyplot as plt

# step function

def step_function(x):
  return np.where(x <= 0, 0, 1)

x = np.linspace(-10, 10)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

```
# sigmoid

x = np.linspace(-10, 10)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y)
plt.show()
```

```
# tanh

x = np.linspace(-5, 5)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x, y)
plt.show()
```

```
# Rectified Linear Unit

x = np.linspace(-5, 5)
y = np.where( x <= 0, 0, x)

plt.plot(x, y)
plt.show()
```

```
#Leaky ReLU

x = np.linspace(-5, 5)
y = np.where( x <= 0, 0.01*x, x)

plt.plot(x, y)
plt.show()
```

```
# Identity function => 회귀문제의 출력층

x = np.linspace(-5, 5)
y = x

plt.plot(x, y)
plt.show()
```

```
# Softmax => 분류문제의 출력층

x = np.array([1, 2, 3])
y = np.exp(x) / np.sum(np.exp(x))

print(y)
```

## 단일 신경망 구현


```
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)
Z = np.zeros((10, 10))
```

```
w_x = 2.5
w_y = 3.0
bias = 0.1
```

```
for i in range(10):
  for j in range(10):
    u = X[i] * w_x + Y[j] * w_y + bias
    y = 1 / (1 + np.exp(-u))
    Z[i, j] = y

plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
